```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from google.colab import files
uploaded = files.upload()
df=pd.read_csv('projek2.csv')

```

```python
plt.figure(figsize=(15,5))

bulan = np.arange(1,13)

plt.bar(bulan, df['1981-1990'], width=0.25, label='1981-1990', color='blue')
plt.bar(bulan + 0.25, df['1991-2000'], width = 0.25, label='1991-2000', color='red')
plt.bar(bulan + 2 * 0.25, df['2001-2010'], width = 0.25, label='2001-2010', color='green')

for i in range(100, 700, 100):
  plt.axhline (y=i, color='black', linestyle='-', linewidth=0.2, zorder=1)

plt.xlabel('Bulan', weight='bold')
plt.ylabel('Curah Hujan (mm)', weight='bold')
plt.legend(loc='upper center', ncols=3, bbox_to_anchor=(0.5,1.2))
plt.xticks(bulan + 0.25, bulan)

plt.show()
```
